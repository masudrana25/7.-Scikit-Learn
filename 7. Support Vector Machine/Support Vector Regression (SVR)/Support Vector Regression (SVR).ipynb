```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

```python
data = pd.read_csv('bangalore house price prediction OHE-data.csv')
data.head()
```

```python
len(data)
```

```python
data2 = pd.DataFrame(data)
data3 = data2.iloc[0:100, :]
len(data3)
```

```python
inputs = data3.drop(['price'], axis =1)
inputs.head()
```

```python
target = data3['price']
target.head()
```

Divide data

```python
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size =0.2)
```

```python
x_train.head()
```

Model Setup

```python
model_rbf = SVR(kernel='rbf')
model_linear = SVR(kernel = 'linear')
model_poly2 = SVR(kernel='poly', degree = 2)
model_poly3 = SVR(kernel='poly', degree = 3)
model_poly4 = SVR(kernel='poly', degree = 4)
model_poly5 = SVR(kernel='poly', degree = 5)
```

```python
model_rbf.fit(x_train, y_train)
model_rbf.score(x_test, y_test)
```

```python
model_linear.fit(x_train, y_train)
model_linear.score(x_test, y_test)
```

```python
model_poly2.fit(x_train, y_train)
model_poly2.score(x_test, y_test)
```

```python
model_poly3.fit(x_train, y_train)
model_poly3.score(x_test, y_test)
```

```python
model_poly4.fit(x_train, y_train)
model_poly4.score(x_test, y_test)
```

```python
model_poly5.fit(x_train, y_train)
model_poly5.score(x_test, y_test)
```

```python
y_predict = model_linear.predict(x_test)
```

Check mse and rmse

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_predict, y_test) ; print(mse)
rmse = np.sqrt(mse) ; print(rmse)
```
