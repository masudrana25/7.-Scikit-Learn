```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
```

```python
df = pd.read_excel('home price.xlsx')
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder( handle_unknown='ignore', sparse_output = False).set_output( transform = 'pandas')
```

```python
ohe_transform = ohe.fit_transform(df[['town']])
ohe_transform
```

```python
tranform_df = pd.concat([df,ohe_transform], axis =1)
tranform_df
```

```python
final_df = tranform_df.drop(['town'], axis=1)
final_df
```

```python
x = final_df.drop(['price'], axis =1)
x
```

```python
y = final_df.price
y
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[2700,1,0,0]])
```
